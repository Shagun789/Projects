```python

gender_dummies = pd.get_dummies(df['Gender'], prefix='', prefix_sep='').astype(int)
gender_dummies = gender_dummies.rename(columns={'Female': 1, 'Male': 0})
```

<!DOCTYPE html>
<html>
<head>
    <title>Bullet Points</title>
</head>
<body>

<ul>
    <li>combining original data set to dummies data set</li>
</ul>

</body>
</html>

```python
df = pd.concat([df, gender_dummies], axis=1)
```

```python
df.head(2)
```

```python
df.rename(columns={1:"Female",0:"Male"},inplace=True)
```
