```python
top_5=df.sort_values(by="Total Cases",ascending=False).head(5)[["State/UTs","Total Cases"]]
bottom_5=df.sort_values(by="Total Cases",ascending=False).tail(5)[["State/UTs","Total Cases"]]
```

```python
bottom_5
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.pie(top_5["Total Cases"],labels=top_5["State/UTs"],autopct="%.2f%%",colors=sns.color_palette("pastel"))
plt.title("Top 5 deadly affected states",color="brown",fontsize=12,fontweight="bold")
plt.subplot(1,2,2)
plt.pie(bottom_5["Total Cases"],labels=bottom_5["State/UTs"],autopct="%.2f%%",colors=sns.color_palette("pastel"))
plt.title("Bottom 5 deadly affecteed states",color="brown",fontsize=12,fontweight="bold");
```

```python
df["Death Ratio (%)"].max()
```

```python
df["Fund_allocation_status"] = df["Death Ratio (%)"].apply(lambda x: "fund allocate" if x >= 1.5 else "fund not allocate")
```

```python

plt.figure(figsize=(12, 5))
plt.plot(df["State/UTs"], df["Active"], label="Active Cases", marker="o")
plt.plot(df["State/UTs"], df["Deaths"], label="Death Cases", marker="^")
plt.xlabel("State/UTs")
plt.ylabel("Number of Cases")
plt.title("Active and Death Cases by State/UTs")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
